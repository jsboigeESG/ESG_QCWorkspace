```python
from AlgorithmImports import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import seaborn as sns


qb = QuantBook()

_eqNames = ['NVDA','ORCL','CSCO','AMD','QCOM']; 
# NVDIA, Oracle, Cisco Systems, Advanced Micro Devices, QUALCOMM Incorporated on sélectionné 5 grandes dans le secteurs technology

_equities = {eq_tickers : qb.AddEquity(
            eq_tickers).Symbol
        for eq_tickers in _eqNames}

start_date = datetime(2020, 6, 1)
end_date = datetime(2024, 6, 1)
histories = {eq : qb.History(_equities[eq] , start_date, end_date, Resolution.Daily) for eq in _equities}

histories['NVDA']
```

```python
symbol_history = {eq: histories[eq].xs(_equities[eq], level='symbol') for eq in _eqNames}
print("Aperçu des données symbol_history :")
for sym in symbol_history:
    print(symbol_history[sym].tail())
```

```python
contracts = {eq: qb.OptionChainProvider.GetOptionContractList(_equities[eq], date(2024,1,19)) for eq in _eqNames}
for eq in _eqNames:
    print(contracts[eq])
```

```python
target_date = { eq: list(set([c.ID.Date.date() for c in contracts[eq]]))[0] for eq in _eqNames}
```

```python
target_date.values()
```

```python
    puts = {eq: [c for c in contracts[eq] if c.ID.Date.date() == target_date[eq] and c.ID.OptionRight == OptionRight.PUT] for eq in _eqNames}
    calls = {eq: [c for c in contracts[eq] if c.ID.Date.date() == target_date[eq] and c.ID.OptionRight == OptionRight.CALL] for eq in _eqNames}
    
```

```python
puts
```

```python
for eq in _eqNames:    
    print("Visualisation des Strikes de l'equity :",eq)
    plt.figure(figsize=(10, 6))
    sns.histplot([p.ID.StrikePrice for p in puts[eq]], bins=20, label="PUTs", color="blue", alpha=0.6)
    sns.histplot([c.ID.StrikePrice for c in calls[eq]], bins=20, label="CALLs", color="red", alpha=0.6)
    plt.title("Distribution des Strikes pour PUTs et CALLs ")
    plt.xlabel("Strike Price")
    plt.ylabel("Nombre de contrats")
    plt.legend()
    plt.show()
```
